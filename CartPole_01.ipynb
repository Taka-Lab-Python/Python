```python
"""
observation=[cart_position(m),cart_velocity(m/sec),pole_angle(rad),pole_angular_velocity(rad/sec)]
reward(報酬)　「1ステップ生き残るごとに +1」　倒れなかった証拠
terminated=True   ポールが倒れすぎた  約 12度（≈ 0.209 rad）
truncated=True    時間制限に達した  500 step に到達 つまり、ポールは倒れていない
"""
```

```python
# CartPole Mainプログラム
print("observation=[cart_position(m),cart_velocity(m/sec),pole_angle(rad),pole_angular_velocity(rad/sec)]","\n")

env = gym.make("CartPole-v1", render_mode="rgb_array")   # 描画の出力形式を指定する引数

deg = 180/3.14
frames = []
observation, info = env.reset()

while True:
    frames.append(env.render())
    move = np.random.choice([0, 1])  # 0：カートを左方向に押す　 1：カートを右方向に押す

    observation, reward, terminated, truncated, info = env.step(move)

    print("observation{}".format(observation))
    print("reward(報酬)={},  terminated={}, truncated={},  move={}".format(reward, terminated, truncated, move),"\n")

    if (terminated or truncated) == True:
       break

    if len(frames) > 501:
       break

env.close()
print("収集したフレーム数:", len(frames))
```

```python
import matplotlib.animation as animation

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)


fig = plt.figure()
plt.axis("off")
im = plt.imshow(frames[0])

def animate(i):
    im.set_array(frames[i])
    return [im]

ani = animation.FuncAnimation(fig, animate, frames=len(frames), interval=100)

# GIFに保存
path = "/content/drive/MyDrive/35_深層強化学習/3-2_CartPole/"
name ="OpenAI-CartPole.mp4"
ani.save(path + name)
print("動画を保存しました⇒", name)
```
